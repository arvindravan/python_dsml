### Agglomerative Clustering

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
import pandas as pd
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.info()
```

```python
# check for the missing vlaue in the DF
s = df.isna().sum()
s[s>0]
```

```python
## chenge the column name as column names has extra values
df.columns = ['CustomerId','Gender','Age','Income','Spending']
```

```python
df.head()
```

```python
# unsupervised learning dont have any Y values we will be having only X values

X = df[['Age','Income','Spending']]
X.head()
```

## Performing preprocessing on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='median')),
                           ('scaler',StandardScaler())])

X_pre = num_pipe.fit_transform(X)
X_pre
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

#### Finding optimal number of clusters with Silhoutee score

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
```

```python
n = list(range(2,16))
n
```

```python
scores=[]
for i in n:
    agc=AgglomerativeClustering(n_clusters=i)
    agc.fit(X_pre)
    lb = agc.labels_
    sh = silhouette_score(X_pre,lb)
    scores.append(sh)
```

```python
scores
```

#### Plot the score to find out max score

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(n,scores)
plt.xlabel('n_cluster')
plt.ylabel('silhouette_score')
plt.title('Optipal number of clusters')
plt.grid()
plt.show()
```

#### Create a aglomerative cluster model with n_clusters=6 as it has highest silhoutee score of 0.42

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
lb = model.labels_
lb
```

#### Give label for X data

```python
X['Label']=lb
```

```python
X.head()
```

#### Show different cluster in data

```python
for i in range(6):
    print(f'Cluster number : {i}')
    display(X[X['Label']==i])
    print('\n=========================================\n')
```

#### Centroids

```python
C = X.groupby(by='Label').mean()
C
```

### Plotly express to visualize cluster

```python
import plotly.express as px
px.scatter_3d(data_frame=X,x='Age',y='Income',z='Spending',color='Label')
```

#### Checking Dendogram

```python
from scipy.spatial import distance_matrix
dm = distance_matrix(X_pre,X_pre)
dm
```

```python
dm.shape
```

```python
from scipy.cluster.hierarchy import linkage
linkage = linkage(dm)
```

```python
from scipy.cluster.hierarchy import dendrogram
Q = dendrogram(linkage)
```
