#### Step 3: univariate analysis

```python
from  matplotlib import pyplot as plt
import seaborn as sb
```

```python
df["species"].value_counts()
```

```python
df["species"].value_counts().plot(kind="bar",title="Count plot for Species")
```

```python
con
```

```python
for i in con:
    sb.histplot(data=df,x=i,kde=True)
    plt.title(f"Histogram for {i}")
    plt.show()
```

```python
for i in con:
    sb.histplot(data=df,x=i,kde=True,hue="species")
    plt.title(f"Histogram for {i}")
    plt.show()
```

#### creating subplot

```python
con
```

```python
plt.figure(figsize=(15,15))
x=1
for i in con:
    plt.subplot(2,2,x)
    x=x+1
    sb.histplot(data=df,x=i,kde=True)
    plt.title(f"Histogram for {i}")
plt.show()
```

```python
plt.figure(figsize=(15,15))
x=1
for i in con:
    plt.subplot(2,2,x)
    x=x+1
    sb.histplot(data=df,x=i,kde=True,hue='species')
    plt.title(f"Histogram for {i}")
plt.show()
```
